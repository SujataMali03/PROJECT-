```python
# Currency Converter Project with Exception & File Handling
# Developed by: Sujata Mali

# Dictionary for exchange rates (Base: USD)
exchange_rates = {
    "USD": 1.0,    #United States Dollar
    "INR": 83.5,   #Indian Rupees
    "EUR": 0.92,    #Euro
    "GBP": 0.78,    #Great British Pound
    "JPY": 151.4,   #Japanese Yen
    "AUD": 1.52,   # Australian Dollar
    "CAD": 1.36,    #Canadian Dollar
    "CNY": 7.23    # Chinese Yuan Renminbi
}

def convert_currency(amount, from_currency, to_currency):
    """Convert the amount from one currency to another."""
    try:
        if from_currency not in exchange_rates or to_currency not in exchange_rates:
            raise ValueError("Invalid currency code entered.")
        converted = (amount / exchange_rates[from_currency]) * exchange_rates[to_currency]
        return converted
    except Exception as e:
        print(f"⚠️ Error during conversion: {e}")
        return None

def save_to_file(from_currency, to_currency, amount, result):
    """Save conversion details to a file."""
    try:
        with open("conversion_history.txt", "a") as file:
            file.write(f"{amount} {from_currency} = {result:.2f} {to_currency}\n")
        print("✅ Conversion saved to 'conversion_history.txt'")
    except Exception as e:
        print(f"⚠️ File handling error: {e}")

def main():     # def main is used to create main function in python
    print("\n💱 Simple Currency Converter (with Exception & File Handling)")
    print("Available Currencies:", ', '.join(exchange_rates.keys()))

    try:
        from_currency = input("Enter currency you have: ").upper()
        to_currency = input("Enter currency you want: ").upper()
        amount = float(input("Enter amount: "))

        result = convert_currency(amount, from_currency, to_currency)
        if result is not None:
            print(f"\n✅ {amount:.2f} {from_currency} = {result:.2f} {to_currency}")
            save_to_file(from_currency, to_currency, amount, result)

    except ValueError:
        print("⚠️ Please enter a valid numeric amount.")
    except Exception as e:
        print(f"⚠️ Unexpected error: {e}")
usd
    finally:
        print("💡 Thank you for using the Currency Converter!")

if __name__ == "__main__":
    main()
```
